```python
# Gerekli kütüphaneleri içe aktarıyoruz
import pandas as pd       # Veri okuma ve veri çerçevesi işlemleri için
import numpy as np        # Sayısal hesaplamalar için
import matplotlib.pyplot as plt  # Grafikler için
import seaborn as sns     # İleri düzey istatistiksel görselleştirme için
import plotly.express as px  # Etkileşimli grafikler için

# Veri setini CSV dosyasından yüklüyoruz
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")

# Veri setinin ilk 5 satırını görüntüleyerek yapısını kontrol ediyoruz
print(data.head())
```

Bu veri seti, Hindistan’daki farklı eyalet ve bölgeler için aylık işsizlik ve istihdam istatistiklerini içermektedir. Her satır, belirli bir bölge ve tarihe ait ölçümleri yansıtmaktadır. Veride, işsizlik oranı yüzdesi, tahmini çalışan sayısı ve işgücüne katılım oranı gibi önemli ekonomik göstergeler yer almakta, ayrıca her bölgenin coğrafi konumu ve koordinat bilgileri de bulunmaktadır. Bu yapı, işsizlik oranlarının zaman içindeki değişimini, bölgeler arası farklılıkları ve işgücüne katılım trendlerini analiz etmeyi mümkün kılmaktadır. Böylece veri seti, ekonomik eğilimleri anlamak ve bölgesel karşılaştırmalar yapmak için oldukça uygun bir kaynak sunmaktadır.

```python
# Her sütundaki eksik değer (NaN) sayısını kontrol ediyoruz
print(data.isnull().sum())
```

Veri setinde yapılan eksik değer kontrolü sonucunda, tüm sütunlarda eksik (NaN) değer bulunmadığı tespit edilmiştir. Bu durum, veri setinin eksiksiz ve güvenilir olduğunu göstermektedir. Dolayısıyla analizlerimizi doğrudan veri üzerinde uygulayabilir ve ek temizleme işlemlerine gerek duymadan istatistiksel ve görselleştirme çalışmalarına geçebiliriz.

```python
# Sütun isimlerini daha okunabilir ve anlaşılır hâle getiriyoruz
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]

# Grafik başlığını belirliyoruz
plt.title("Indian Unemployment")

# Bölgeler (Region) bazında tahmini istihdam sayılarını gösteren histogram çiziyoruz
sns.histplot(x="Estimated Employed", hue="Region", data=data)

# Grafiği göster
plt.show()
```

Bu grafikte bölgeler bazında tahmini istihdam dağılımları görülmektedir. Kuzey (North) ve Güney (South) bölgelerinin istihdam düzeyleri diğer bölgelere kıyasla daha yoğun gözlemlenmektedir. Bu da bu bölgelerde iş gücü piyasasının daha geniş ve aktif olduğunu göstermektedir. Batı (West) ve Doğu (East) bölgelerinde ise istihdam sayıları görece daha sınırlı kalmıştır. Dolayısıyla iş gücünün ülke genelinde eşit dağılmadığı, bazı bölgelerin istihdam açısından daha avantajlı olduğu anlaşılmaktadır.

```python
# Grafik boyutunu ayarlıyoruz
plt.figure(figsize=(12, 10))

# Grafik başlığını belirliyoruz
plt.title("Indian Unemployment")

# Bölgeler (Region) bazında tahmini işsizlik oranlarını gösteren histogram çiziyoruz
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)

# Grafiği göster
plt.show()
```

Bu grafikte işsizlik oranlarının dağılımı incelenmektedir. Çoğu bölgedeki işsizlik oranları %0 ile %20 aralığında yoğunlaşmaktadır. Ancak Güney (South) ve Doğu (East) bölgelerinde zaman zaman %40’ın üzerinde işsizlik oranlarına rastlanmaktadır. Bu durum, bazı bölgelerde yapısal sorunlar ve istihdam yetersizliği olduğunu düşündürmektedir. Ayrıca işsizlik oranlarının düşük olduğu bölgelerde bile, istihdamın kalitesi ve iş gücü katılım oranı gibi faktörlerin dikkate alınması gerektiği ortaya çıkmaktadır.
